```python
import numpy as np
import matplotlib.pyplot as plt
```

# Uniformly charged sphere

```python
r = np.linspace(0, 10, 1000)
E = np.piecewise(r, [r < 1, r >= 1], [lambda r: r, lambda r: 1 / r**2])

plt.figure(figsize=(8, 6))
plt.plot(r, E, label=r'$E(r):\ r < 1 \rightarrow r,\ r \geq 1 \rightarrow 1/r^2$', color='blue', linewidth=2)

# Aesthetic improvements
plt.title('Plot of $E(r)$', fontsize=14)
plt.xlabel('$r$', fontsize=12)
plt.ylabel('$E(r)$', fontsize=12)
plt.axvline(x=1, color='gray', linestyle='--', linewidth=1, label='$r=1$ (radius of the uniformly charged sphere)')
plt.legend(fontsize=10, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig("_Media/uniformly-charged-sphere.png")
# Show the plot
plt.show()
```

# Shell

```python
r = np.linspace(0, 10, 1000)
E = np.piecewise(r, [r < 1, r >= 1], [lambda r: 0, lambda r: 1 / r**2])

plt.figure(figsize=(8, 6))
plt.plot(r, E, label=r'$E(r):\ r < 1 \rightarrow 0,\ r \geq 1 \rightarrow 1/r^2$', color='blue', linewidth=2)

# Aesthetic improvements
plt.title('Plot of $E(r)$', fontsize=14)
plt.xlabel('$r$', fontsize=12)
plt.ylabel('$E(r)$', fontsize=12)
plt.axvline(x=1, color='gray', linestyle='--', linewidth=1, label='$r=1$ (radius of the uniformly charged sphere)')
plt.legend(fontsize=10, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.tight_layout()

plt.savefig("_Media/shell.png")
# Show the plot
plt.show()
```
